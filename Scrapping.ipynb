2) Loading required website for data extraction

```python
from urllib.request import urlopen
url = "https://gate.appliedroots.com/upcomingexams"
data = urlopen(url)
html = data.read()
print(html)
data.close()
```

3) extracting required data

```python
from bs4 import BeautifulSoup as s
soup = s(html,'html.parser')
print(soup)
```

```python
table = soup.find('table',{})
print(table)
```

```python
header = table.findAll('th')[1:]
print(header)
```

```python
header_text = [txt.text[1:] for txt in header]
print(header_text)
```

```python
data = table.findAll('tr')[2:]
table_rows = []
for row in data:
    attributes = row.findAll('td')[:-4]
    current_row = []
    for at in attributes:
        current_row.append(at.text[1:-1])
    attributes = row.findAll('td')[3:]
    current_row.append(attributes[0].text[1:])
    current_row.append(attributes[1].text[1:])
    current_row.append(attributes[2].text[1:])
    current_row.append(attributes[3].text[1:-1])
    table_rows.append(current_row)
print(table_rows)
```

4) CSV

```python
fileName = 'exams.csv'
with open(fileName,'w',encoding='utf-8') as f:
    header_string = ','.join(header_text)
    header_string += '\n'
    f.write(header_string)
    for row in table_rows:
        row_string = ','.join(row)
        row_string += '\n'
        f.write(row_string)    
```
